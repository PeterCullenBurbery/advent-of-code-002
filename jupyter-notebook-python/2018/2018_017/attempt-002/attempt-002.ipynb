```python
def create_grid_with_headers_and_simulate_water(input_file_path, output_file_path):
    with open(input_file_path, 'r') as file:
        clay_coordinates = file.readlines()

    # Parse the input and record clay positions
    grid = {}
    for line in clay_coordinates:
        line = line.strip()
        if line.startswith('x'):
            x, y_range = line.split(', ')
            x = int(x[2:])
            y_start, y_end = map(int, y_range[2:].split('..'))
            for y in range(y_start, y_end + 1):
                grid[(x, y)] = '#'
        elif line.startswith('y'):
            y, x_range = line.split(', ')
            y = int(y[2:])
            x_start, x_end = map(int, x_range[2:].split('..'))
            for x in range(x_start, x_end + 1):
                grid[(x, y)] = '#'

    # Define bounds
    min_x = 1
    max_x = max(x for x, _ in grid.keys()) + 1  # Include max_x + 1 for clarity
    max_y = max(y for _, y in grid.keys())  # Bottommost clay coordinate

    # Generate x-coordinate headers
    headers = []
    for i in range(3):  # Generate three rows for headers (hundreds, tens, ones)
        header_row = []
        for x in range(min_x, max_x + 1):
            header_row.append(str((x // (10**(2 - i))) % 10))
        headers.append("".join(header_row))

    # Create the grid
    result_grid = []
    for y in range(0, max_y + 1):  # Ensure grid starts at y=0
        row = []
        for x in range(min_x, max_x + 1):
            if (x, y) == (500, 0):  # Place the source
                row.append('+')
            else:
                row.append(grid.get((x, y), '.'))  # Default to sand
        result_grid.append(row)

    # Simulate water flow downwards
    source_x = 500 - min_x  # Adjusted for grid indexing
    for y in range(1, max_y + 1):  # Start below the spring
        if result_grid[y][source_x] == '#':  # Stop at clay
            break
        result_grid[y][source_x] = '|'

    # Convert enclosed areas to settled water (~)
    for y in range(max_y + 1):  # Iterate through rows
        row = result_grid[y]
        start = 0
        while start < len(row):
            # Look for a pattern: #....|....#
            if row[start] == '#':
                end = start + 1
                while end < len(row) and row[end] != '#':
                    end += 1
                if end < len(row) and '|' in row[start + 1:end]:
                    # Replace | and . with ~ in the enclosed area
                    for i in range(start + 1, end):
                        if row[i] in '|.':
                            row[i] = '~'
            start += 1

    # Write the headers and grid to the output file
    with open(output_file_path, 'w') as output_file:
        for header_row in headers:
            output_file.write(header_row + '\n')
        for row in result_grid:
            output_file.write(''.join(row) + '\n')
```
